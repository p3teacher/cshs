```python
# ASSNP-02.py
# 2021-11-15
#
# 將" "分界改為"," ; 產生 dem-all-2.csv file
#
# run on myBinder.org JupyterLab
# 程式檔 path is '/home/jovyan/SSNP3D/2_dir'
# 資料檔 path is '/home/jovyan/SSNP3D/2_dir/data'
# 圖形檔 path is '/home/jovyan/SSNP3D/2_dir/graphic'
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
import csv
import numpy as np
import pprint
#
path_data    = './data/'
#
# 逐筆讀入資料 dem-all-1.grd (149328 row)
# Read from the file into the array data(:,:)
# array data shape is (149328, 3)
#
data1 = np.loadtxt(path_data + 'dem-all-1.grd',delimiter=' ',dtype='float')
#
i = 0
for row in data1 :
    i = i + 1
    # 顯示前五筆資料
    if i < 6 :
        pprint.pprint (row)
        for field in row :
            print (field)
        # end for
        print ('\n')
    # end if
# end for
#
float_To_int = np.vectorize(np.int32)
# Columns.
# data[..., 0] is array of float number
# data[..., 1] is array of float number
# data[..., 2] is array of float number
#
# for numpy -->  data[..., 1] 等價于 data[:, 1]
#
# 台灣橫麥卡托二度分帶投影座標系統（TM2）
#
TM2_X = float_To_int(data1[..., 0]) # 二度分帶 X座標
TM2_Y = float_To_int(data1[..., 1]) # 二度分帶 Y座標
TM2_Z = data1[..., 2]               # DEM file z value
#
i = 0
j = 0
for zValue in TM2_Z :
    i = i + 1
    # 顯示前五筆資料
    if i < 6 :
        print (zValue)
    # end if
    if zValue < 0 :
        j = j + 1
    # end if
# end for
#
print ('\n')
print (TM2_X.size) # 149328
print (TM2_Y.size) # 149328
print (TM2_Z.size) # 149328
print ('zValue < 0 筆數有',j)
print ('\n')
#
i = 0
dem_list = list(zip(TM2_X, TM2_Y, TM2_Z))
#
# 顯示前五筆資料
print (dem_list[0:5]) # begin at 0, total 5 items
print ('\n')
#
with open(path_data + 'dem-all-2.csv','wt', newline='', encoding='utf-8') as outfile :
    dem = csv.writer(outfile)
    dem.writerows(dem_list)
# end with    
#
print ('Done')
```
